```python
f = open('keys/.openai_api_key.txt')
OPEN_API_KEY=f.read()
```

#### Using the llm model

```python
from langchain_openai.llms import OpenAI
llm=OpenAI( api_key=OPEN_API_KEY,model="gpt-4o-mini",temperature=0)
```

```python
prompt="How many states are there in India"
```

```python
print(llm.invoke(prompt))
```
